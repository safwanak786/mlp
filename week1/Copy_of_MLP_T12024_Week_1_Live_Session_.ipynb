## Q.19 Insert a dummy column 'sampleDate', which signifies date of on which the relevant medical tests were performed.
- Randomly generate dates.
- What is type of this new column?

```python
def randomDate(dummy=0):
  year = np.random.randint(2010, 2025)
  month = np.random.randint(1, 13)
  date = np.random.randint(1, 29)

  return f'{year}-{month}-{date}'

# np.random.seed(0)

randomDate()
```

```python
# df['sampleDate'] = df.randomDate()
df['sampleDate'] = df['age'].apply(randomDate)
```

```python
df['sampleDate'] = [ randomDate() for i in range(len(df))]
```
